## Boto3 Demo

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
import boto3
```

```python
s3_client = boto3.client('s3', region_name='us-east-2')
```

```python
buckets = s3_client.list_buckets()
buckets
```

```python
buckets.keys()
```

```python
for bucket in buckets.get('Buckets', 'none found'):
    if 'techcatalyst' in bucket.get('Name'):
        print(bucket.get('Name'))
```

```python
objects_list = s3_client.list_objects_v2(Bucket='techcatalyst-raw')
objects_list.keys()
```

```python
objects_list['Contents']
```

```python
# get all csvs
for object in objects_list.get('Contents', None):
    if object.get('Key').endswith('.csv'):
        print(object.get('Key'))
```

```python
# OR do this
for object in objects_list.get('Contents', None):
    output = object.get('Key').split('.')
    if len(output) > 1:
        if output[1] == 'csv':
            print(object.get('Key'))
```
